```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/insurance.csv')
df.head()
```

```python
df.describe()
```

```python
age = df['age'].values
expenses = df['expenses'].values
df['region'] = df['region'].astype('category')

data_list = [age, expenses, df['region']]
matrix = df[['age', 'expenses']].values
```

```python

# b) Generate a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(age, expenses, c='blue', marker='+')
plt.title('Age vs Expenses')
plt.xlabel('Age (years)')
plt.ylabel('Expenses (USD)')
plt.grid(True)
for i in range(0, len(age), 50):
    plt.text(age[i], expenses[i], df['region'].iloc[i], fontsize=9)

plt.show()
```

```python
filtered_df = df[df['region'] == 'southeast']

plt.figure(figsize=(10, 6))
plt.scatter(filtered_df['age'], filtered_df['expenses'], c='red', marker='+')
plt.title('Age vs Expenses (Filtered: Southeast Region)')
plt.xlabel('Age (years)')
plt.ylabel('Expenses (USD)')
plt.grid(True)

plt.show()
```

```python
smoking_data = df[df['smoker'] == 'yes']

plt.figure(figsize=(10, 6))
plt.hist(smoking_data['age'], bins=10, color='purple', edgecolor='black')
plt.title('Age Group Distribution of Smokers')
plt.xlabel('Age (years)')
plt.ylabel('Number of Smokers')
plt.grid(True)

plt.show()


smoking_data = df[df['smoker'] == 'yes']
plt.figure(figsize=(10, 6))
plt.scatter(smoking_data['age'], np.arange(len(smoking_data)), c='purple', marker='o', edgecolor='black')
plt.title('Age Group Distribution of Smokers')
plt.xlabel('Age (years)')
plt.ylabel('Index of Smokers')
plt.grid(True)

plt.show()
```

```python
smoking_data = df[(df['smoker'] == 'yes') & (df['age'] >= 20) & (df['age'] <= 40)]
plt.figure(figsize=(10, 6))
plt.scatter(smoking_data['age'], np.arange(len(smoking_data)), c='purple', marker='o', edgecolor='black')
plt.title('Age Group Distribution of Smokers (Age 20-40)')
plt.xlabel('Age (years)')
plt.ylabel('Index of Smokers')
plt.grid(True)

plt.show()
```
